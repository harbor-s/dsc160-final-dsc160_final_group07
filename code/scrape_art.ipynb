```python
from bs4 import BeautifulSoup
import os
import requests
import numpy as np
from skimage import io
import skimage
import pandas as pd
import os
import matplotlib.pyplot as plt
import scipy.misc
from skimage import data
from skimage.color import rgb2hsv
```

```python
DATA_DIR = 'data/'
ARTIST_URL = 'https://www.wikiart.org/en/{artist}/all-works/text-list'
PAINTING_URL = 'https://www.wikiart.org{painting_path}'
```

```python
if not os.path.exists(DATA_DIR):
    os.makedirs(DATA_DIR)
```

```python
artist_names = ['jacob-lawrence', 'aaron-douglas', 'clementine-hunter', 'jean-michel-basquiat', 'kerry-james-marshal', 'romare-bearden', 'alma-woodsey-thomas', 'david-hammons', 'henry-ossawa-tanner', 'william-h-johnson']

def getArt(artist_name):
    
    url_query = ARTIST_URL.format(artist=artist_name)
    artist_page = requests.get(url_query)


    # check for request error
    try:
        artist_page.raise_for_status()
    except requests.exceptions.HTTPError as e:
        print("Error trying to retrieve {}".format(artist_page.url))
        raise e

    soup = BeautifulSoup(artist_page.text, 'lxml')

    IMAGE_DIR = os.path.join(DATA_DIR, artist_name)
    if not os.path.exists(IMAGE_DIR):
        os.makedirs(IMAGE_DIR)
        
    painting_paths = []

    # retreive all rows in painting-list
    for li in soup.find_all('li', {'class': 'painting-list-text-row'}):
        # retrieve all links in the current row
        for link in li.find_all('a'):
            href = link.get('href')
            # store in dictionary
            painting_paths.append(href)
          
        
    def download_and_save(painting_url):
        r_painting_page = requests.get(painting_url)
        soup = BeautifulSoup(r_painting_page.text, 'lxml')
        for img in soup.find_all('img', {'class': 'ms-zoom-cursor'}):
            img_url = img['src']
            img_url = img_url.split('!')[0]
            filename = img_url.split('/')[-1]

            outfile = os.path.join(IMAGE_DIR, filename)                       
            if not os.path.exists(outfile):                        
                print("downloading {}: {}".format(filename, img_url))
                r = requests.get(img_url, outfile)
                with open(outfile, 'wb') as f:
                    f.write(r.content)
            else:
                #print("skipping {}".format(filename))
                pass
        
        
    for path in painting_paths:
        painting_path = PAINTING_URL.format(painting_path=path)
        download_and_save(painting_path)
    
```

```python
for artist in artist_names:
    getArt(artist)
```

```python
root='data'

#Set your own PATH 
IMAGE_SIZE = 128
IMAGE_CHANNELS = 3
ARTIST_DIR = 'data/'
SAMPLE_SIZE = 20

# Defining image dir path. Change this if you have different directory
artist_path = ARTIST_DIR 

training_data = []

# Iterating over the images inside the directory and resizing them using
# Pillow's resize method.
print('resizing...')

for artist_dir in os.listdir(artist_path):
    images_path = os.path.join(artist_path, artist_dir)
    if len(os.listdir(images_path)) < SAMPLE_SIZE:
        random_images = os.listdir(images_path)
    else:
        random_images = random.sample(os.listdir(images_path), SAMPLE_SIZE)
    
    for filename in random_images:
        path = os.path.join(images_path, filename)
        image = Image.open(path).convert('RGB').resize((IMAGE_SIZE, IMAGE_SIZE), Image.ANTIALIAS)
        training_data.append(np.asarray(image))

training_data = np.reshape(
    training_data, (-1, IMAGE_SIZE, IMAGE_SIZE, IMAGE_CHANNELS))
training_data = training_data / 127.5 - 1

print(training_data.shape)

print('saving file...')
np.save('small_dataset.npy', training_data)
```
